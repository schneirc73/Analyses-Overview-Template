# High Level View

```python
Domain_Count=df.groupby('Domain')['Project ID'].count().reset_index()

sns.barplot(data = Domain_Count, x='Domain', y='Project ID')
plt.title("Project across all Domains")
plt.xlabel("Project Domains")
plt.ylabel("Project Count")
plt.show()
```

```python
Analysis_Count=df.groupby('Type of Analysis')['Project ID'].count().reset_index()

sns.barplot(data = Analysis_Count, x='Type of Analysis', y='Project ID')
plt.title("Types of Analyses Utilized Across All Projects")
plt.xlabel("Type of Analysis")
plt.ylabel("Analysis Count")
plt.show()
```

```python
Outcome_Count=df.groupby('Outcome')['Project ID'].count().reset_index()

sns.barplot(data = Outcome_Count, x='Outcome', y='Project ID')
plt.title("Count of Project Outcomes Across All Projects")
plt.xlabel("Outcomes")
plt.ylabel("Outcome Count")
plt.show()
```

```python
Tools_Count=df.groupby('Tools')['Project ID'].count().reset_index()

sns.barplot(data = Tools_Count, x='Tools', y='Project ID')
plt.title("Tool Usage Across All Projects")
plt.xlabel("Tool")
plt.ylabel("Tool Utilization Count")
plt.show()
```

# In-Depth View

```python
for domain in df['Domain'].unique():
    print(domain)
    temp = df[df['Domain'] == domain]
    temp_group = temp.groupby('Tools')['Project ID'].count().reset_index()
    plt.pie(temp_group['Project ID'], labels=temp_group['Tools'], autopct='%.0f%%')
    plt.title(f"Tool Usage within {domain} Project Area")
    plt.show()   
```

```python
for domain in df['Domain'].unique():
    print(domain)
    temp = df[df['Domain'] == domain]
    temp_group = temp.groupby('Project Members')['Project ID'].count().reset_index()
    plt.pie(temp_group['Project ID'], labels=temp_group['Project Members'], autopct='%.0f%%')
    plt.title(f"Team Members working in {domain} Project Area")
    plt.show()   
```
